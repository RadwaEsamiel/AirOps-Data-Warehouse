```python
def generate_insert_statement(num_runways_list, terminal_count_list, num_gates_list, atc_capacity_list, parking_stands_list, guest_capacity_list):
    with open('Populate_Airport.SQL', 'w') as data:
        for i in range(len(airports)):
            airport = airports[i]
            airport_id = generate_airport_id()
            airport_name = airport["Airport_Name"]
            country = airport["Country"]
            city = airport["City"]
            latitude = airport["Latitude"]
            longitude = airport["Longitude"]
            region = airport["Region"]
            num_runways = random.choice(num_runways_list)
            terminal_count = random.choice(terminal_count_list)
            num_gates = random.choice(num_gates_list)
            atc_capacity = random.choice(atc_capacity_list)
            aircraft_parking_stands = random.choice(parking_stands_list)
            guest_capacity = random.choice(guest_capacity_list)

            insert_statement = (
                f"INSERT INTO Airport_Dimension "
                f"VALUES ('1', '{airport_id}', '{airport_name}', '{country}', '{city}', {latitude}, {longitude}, "
                f"'{region}', {num_runways}, {terminal_count}, {num_gates}, {atc_capacity}, {aircraft_parking_stands}, {guest_capacity});\n"
            )
            data.write(insert_statement)


```
