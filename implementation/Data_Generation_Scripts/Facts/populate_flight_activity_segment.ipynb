```python
import random
from datetime import datetime, timedelta
# Define ranges for each item
passenger_key_range = (9929, 19854)
passenger_profile_range = (1, 1200)
start_date = datetime(2014, 1, 1)
end_date = datetime(2023, 12, 31)
time_range = (1, 1440)
airport_range = (1, 37)
booking_channel_range = (1, 100)

aircraft_range = (1, 100)
class_of_service_range = (1, 64)
fare_basis_range = (1, 500)
ticket_number_range = (10000, 9999999)
segment_sequence_number_range = (1, 100000)
flight_number_range = (1000, 99999)

confirmation_number_dd_range = range(100000, 1000000) 
base_fare_revenue_range = range(1000, 10000000) 
overnight_stay_range = range(1,5)
airport_tax_range = range(100, 1000)  
government_tax_range = range(100, 1000)
passenger_facility_charges_range = range(10,1000) 
baggage_charges_range = range(100, 1000)  
upgrade_fees_range = range(100, 1000)  
transaction_fees_range = range(100, 1000)  
segment_miles_flown_range = range(100, 1000)  
segment_miles_earned_range = range(100, 1000)  
trip_duration_range = range(1, 30)  
overnight_stay_duration_range = range(1, 10)  
delay_duration_range = range(1, 12)  

```

```python
def random_decimal(min_value, max_value):
    return round(random.uniform(min_value, max_value), 2)

# Function to generate random insert statement
def generate_insert():
    random_date = start_date + timedelta(days=random.randint(0, (end_date - start_date).days))
    scheduled_departure_date_key = int(random_date.strftime('%Y%m%d'))
    scheduled_departure_time_key = random.randint(time_range[0], time_range[1])

    # Generate the second random date with an offset of up to 1 day
    random_offset = timedelta(days=random.uniform(0, 1))
    random_date2 = random_date + random_offset
    actual_departure_date_key = int(random_date2.strftime('%Y%m%d'))
    actual_departure_time_key = random.randint(time_range[0], time_range[1])

    insert_statement = "INSERT INTO fact_flight_activity_segment VALUES ("
    insert_statement += str(scheduled_departure_date_key) + ", "  # Scheduled_Departure_Date_Key
    insert_statement += str(scheduled_departure_time_key) + ", "  # Scheduled_Departure_Time_Key
    insert_statement += str(actual_departure_date_key) + ", "  # Actual_Departure_Date_Key
    insert_statement += str(actual_departure_time_key) + ", "  # Actual_Departure_Time_Key
    insert_statement += str(random.randint(passenger_key_range[0], passenger_key_range[1])) + ", "  # Passenger_Key
    insert_statement += str(random.randint(passenger_profile_range[0], passenger_profile_range[1])) + ", "  # Passenger_Profile_Key
    insert_statement += str(random.randint(airport_range[0], airport_range[1])) + ", "  # Segment_Origin_Airport_Key
    insert_statement += str(random.randint(airport_range[0], airport_range[1])) + ", "  # Segment_Destination_Airport_Key
    insert_statement += str(random.randint(aircraft_range[0], aircraft_range[1])) + ", "  # Aircraft_Key
    insert_statement += str(random.randint(class_of_service_range[0], class_of_service_range[1])) + ", "  # Class_of_Service_Flown_Key
    insert_statement += str(random.randint(fare_basis_range[0], fare_basis_range[1])) + ", "  # Fare_Basis_Key
    insert_statement += str(random.randint(booking_channel_range[0], booking_channel_range[1])) + ", "  # Booking_Channel_Key
    insert_statement += str(random.choice(confirmation_number_dd_range)) + ", "  # Confirmation_Number_DD
    insert_statement += str(random.randint(ticket_number_range[0], ticket_number_range[1])) + ", "  # Ticket_Number_DD
    insert_statement += str(random.randint(segment_sequence_number_range[0], segment_sequence_number_range[1])) + ", "  # Segment_Sequence_Number_DD
    insert_statement += str(random.randint(flight_number_range[0], flight_number_range[1])) + ", "  # Flight_Number_DD
    insert_statement += str(random.choice(base_fare_revenue_range)) + ", "  # Base_Fare_Revenue
    insert_statement += str(random.choice(passenger_facility_charges_range)) + ", "  # Passenger_Facility_Charges
    insert_statement += str(random.choice(airport_tax_range)) + ", "  # Airport_Tax
    insert_statement += str(random.choice(government_tax_range)) + ", "  # Government_Tax
    insert_statement += str(random.choice(baggage_charges_range)) + ", "  # Baggage_Charges
    insert_statement += str(random.choice(upgrade_fees_range)) + ", "  # Upgrade_Fees
    insert_statement += str(random.choice(transaction_fees_range)) + ", "  # Transaction_Fees
    insert_statement += str(random.choice(segment_miles_flown_range)) + ", "  # Segment_Miles_Flown
    insert_statement += str(random.choice(segment_miles_earned_range))  # Segment_Miles_Earned
    insert_statement += ");\n"
    return insert_statement

# Write insert statements to file
with open('populate_segment_level_flight_activity_fact.sql', 'w') as data:
    for _ in range(50000):
        insert_statement = generate_insert()
        try:
            data.write(insert_statement)
        except:
            continue
```
